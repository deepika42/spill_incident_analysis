```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://dnaik2awsbucket.s3.amazonaws.com/Spill_Incidents_RAW.csv")
print(df.shape)
# df.insert(loc=0, column='Row Number', value=np.arange(len(df)))
df.head()
```

```python
df.isnull().sum() # Checking for null values
```

```python
df.duplicated() # Checking for duplicates
```

```python
df[['Spill Date', 'Received Date', 'Close Date']] = df[['Spill Date', 'Received Date', 'Close Date']].apply(pd.to_datetime)

df = df[['Spill Number', 'Street 1', 'Locality', 'County', 'Spill Date', 'Received Date', 'Contributing Factor','Waterbody', 'Close Date', 'Source', 'Material Name', 
         'Material Family', 'Quantity', 'Units', 'Recovered']]

df.dropna(subset=['Spill Date'], inplace=True)
df = df[df['Source']!='Missing Code in Old Data - Must be fixed']

df = df.reset_index(drop=True)

df = df[df['Spill Date'] >= '1984-01-01']

df.head()
```

```python
df['Waterbody'].value_counts()
df['Waterbody'] = df['Waterbody'].str.replace(r'[^A-Za-z\s]', '', regex=True)
```

```python
df = df.sort_values(by='Spill Date')
df.set_index('Spill Date', inplace=True)
df.head()

df.to_csv('Spill_Incidents.csv')
```
